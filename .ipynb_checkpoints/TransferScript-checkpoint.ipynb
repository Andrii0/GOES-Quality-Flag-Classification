```python
import string
import re
import os
import os.path as op
from subprocess import Popen
```

```python
# PATH_IN = z
# PATH_OUT = 'fall_npy'
PATH_IN = '/sharedData2/scratch/fall_npy'
PATH_OUT = 'fall_npy'
```

```python
dirs = os.listdir( PATH_IN )
```

```python
for file in dirs:
    #year = file[27:31]
    day = file[31:34]
    
    inFile = op.join(PATH_IN,file)
    
    # folder = op.join(PATH_IN,year)
    # folder = op.join(folder,day)
    
    folder = op.join(PATH_OUT,day)
    
    outFile = op.join(folder, file)
    
    if op.exists(outFile):
        continue
    
    if op.exists(folder):
        pass
    else:
        cmd = [ " ".join(['mkdir',folder]) ]
        pid = Popen(cmd, shell=True)
        pid.communicate()
        
    cmd = [ " ".join(['mv',inFile,outFile]) ]
    pid = Popen(cmd, shell=True)
    pid.communicate()
    
```

```python
dirs = os.listdir( PATH_IN )
```

```python
for file in dirs:
    if 'ACM' not in file:
        #year = file[27:31]
        day = file[31:34]

        inFile = op.join(PATH_IN,file)

        # folder = op.join(PATH_IN,year)
        # folder = op.join(folder,day)

        folder = op.join(PATH_OUT,day)

        outFile = op.join(folder, file[:38])

        if op.exists(outFile):
            continue

        if op.exists(folder):
            pass
        else:
            cmd = [ " ".join(['mkdir',folder]) ]
            pid = Popen(cmd, shell=True)
            pid.communicate()

        cmd = [ " ".join(['cp',inFile,outFile]) ]
        pid = Popen(cmd, shell=True)
        pid.communicate()
        print(file)
    else:
        continue
```
