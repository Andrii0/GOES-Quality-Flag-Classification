```python
import matplotlib.pyplot as plt
import h5py
import numpy as np
```

```python
input_path = '/Users/aleex/DSE/Machine Learning/Project/label/'

file = input_path + 's20190980000.h5'

with h5py.File(file, 'r') as f:
    label = f['label'][()]
    data = f['data'][()]

data = data.reshape((1001, 401, 4))
label = label.reshape((1001, 401))
```

```python
fig, axs = plt.subplots(1, 1, figsize=(12, 16))
fig.subplots_adjust(wspace=0.5)

img = axs.imshow(label, interpolation='none', vmin = 0, vmax=5, cmap='gray')
axs.set_title('s20190980000 new label')
fig.colorbar(img, ax=axs, fraction=0.1)
```

```python
fig, axs = plt.subplots(1, 5, figsize=(15, 25))
fig.subplots_adjust(wspace=0.5)

img0 = axs[0].imshow(data[:,:,0],interpolation='none')
axs[0].set_title('band 07')
fig.colorbar(img0, ax=axs[0], fraction=0.1)

img1 = axs[1].imshow(data[:,:,1],interpolation='none')
axs[1].set_title('band 08')
fig.colorbar(img1, ax=axs[1], fraction=0.1)

img2 = axs[2].imshow(data[:,:,2],interpolation='none')
axs[2].set_title('band 09')
fig.colorbar(img2, ax=axs[2], fraction=0.1)

img3 = axs[3].imshow(data[:,:,3],interpolation='none')
axs[3].set_title('band 10')
fig.colorbar(img3, ax=axs[3], fraction=0.1)

img4 = axs[4].imshow(label, interpolation='none', vmin = 0, vmax=5, cmap='gray')
axs[4].set_title('s20190980000 new label')
fig.colorbar(img4, ax=axs[4], fraction=0.1)
```
