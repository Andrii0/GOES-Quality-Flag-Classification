```python
import string
import re
import os
import os.path as op
from subprocess import Popen
```

```python
PATH_DQF= '../../goesdata/DQF'
PATH_DATA = '../../goesdata/fall_npy'
```

```python
"OR_ABI-L1b-RadF-M6C07_G16_s20192210000"[:38]
```

```python
dirsDQF = os.listdir( PATH_DQF)
dirsDATA = os.listdir( PATH_DATA )
```

```python
for day in dirsDATA:
    PATH_IN = op.join(PATH_DATA,day)
    
    for file in os.listdir( PATH_IN ):
        trunc = file[:38]
        inFile =  op.join(PATH_IN,file) 
        outFile = op.join(PATH_IN,trunc)

        if op.exists(outFile):
            continue

        else:
            cmd = [ " ".join(['mv', inFile, outFile]) ]
            print(cmd)
            pid = Popen(cmd, shell=True)
            pid.communicate()

    
```
